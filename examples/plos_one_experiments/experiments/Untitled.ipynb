```python
import numpy as np

r2 = np.genfromtxt('stats_profile_tnqvm_2_REDO_rounds.csv', delimiter=',', names=['r','n','m'])
r4 = np.genfromtxt('stats_profile_tnqvm_4_REDO_rounds.csv', delimiter=',', names=['r','n','m'])
r6 = np.genfromtxt('stats_profile_tnqvm_6_rounds.csv', delimiter=',', names=['r','n','m'])
r8 = np.genfromtxt('stats_profile_tnqvm_8_rounds.csv', delimiter=',', names=['r','n','m'])
r10 = np.genfromtxt('stats_profile_tnqvm_10_rounds-Copy3.csv', delimiter=',', names=['r','n','m'])

def chunks(l, n):
    """Yield successive n-sized chunks from l."""
    for i in range(0, len(l), n):
        yield l[i:i + n]

r2_avg = [np.mean(c) for c in list(chunks(r2['m'], 10))]
r4_avg = [np.mean(c) for c in list(chunks(r4['m'], 10))]
r6_avg = [np.mean(c) for c in list(chunks(r6['m'], 10))]
r8_avg = [np.mean(c) for c in list(chunks(r8['m'], 10))]
r10_avg = [np.mean(c) for c in list(chunks(r10['m'], 10))]

r2_std = [np.std(c) for c in list(chunks(r2['m'], 10))]
r4_std = [np.std(c) for c in list(chunks(r4['m'], 10))]
r6_std = [np.std(c) for c in list(chunks(r6['m'], 10))]
r8_std = [np.std(c) for c in list(chunks(r8['m'], 10))]
r10_std = [np.std(c) for c in list(chunks(r10['m'], 10))]

ns = [5,10,15,20,25,30,35,40,45,50,55,60,65,70,75,80,85]

import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize=(15,10))
plt.rc('font', family='serif', serif='cm10')
plt.rc('text', usetex=True)
plt.rcParams['text.latex.preamble'] = [r'\boldmath']
plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)

plt.rc('lines', lw=2)
plt.rc('axes', linewidth=2)
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.xlabel(r'Number of Qubits', fontsize=20)
plt.ylabel(r'Memory (MB)', fontsize=20)
ax = fig.add_subplot(1, 1, 1)
ax.set_yscale("log", nonposy='clip')
ax.errorbar(ns, r2_avg, yerr=r2_std, capsize=5, color='b', label='2 Rounds')
ax.errorbar(ns, r4_avg, yerr=r4_std, capsize=5, color='r', label='4 Rounds')
ax.errorbar(ns, r6_avg, yerr=r6_std, capsize=5, color='g', label='6 Rounds')
plt.errorbar(ns, r8_avg, yerr=r8_std, capsize=5, color='m', label='8 Rounds')
plt.errorbar([5,10,15,20,25,30,35,40,45,50], r10_avg, yerr=r10_std, capsize=5, color='y', label='10 Rounds')

plt.legend(fontsize=20)
plt.savefig("tnqvm_memory_profiling.pdf",bbox_inches='tight')
plt.show()
```
