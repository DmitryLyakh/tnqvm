```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("build/energy_vs_theta.csv",'r') as f:
    labels = f.readline().strip().replace(" ", "").split(',')
    del labels[0]
print(labels)
energy_vs_theta = np.loadtxt("build/energy_vs_theta.csv", delimiter=',', skiprows=1)
ax = plt.gca()
ax.cla()

for i in range(0,energy_vs_theta.shape[1]-1):
    ax.plot(energy_vs_theta[:,0], energy_vs_theta[:,i+1], label=r"$\left<%s\right>$" % labels[i])
ax.legend()
plt.show()
```
